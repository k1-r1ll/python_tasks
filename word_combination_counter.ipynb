# Word Combinaton Count

**Main task:** to check, how many words one can get from the word ‘коммуникация’ (communication).

For the task we will use some orphographic dictionary in ‘txt.’ format. First of all, we import regular expression module and create an empty list for the words after clearing the basic dictionary.

```python
import re
clear_basic_data = list()
```

Then we added a dictionary. Opened it and cleared all the lines. If the line was not empty, we would take its second element (as some of the basic words repeats in one line because of flexion issues) and clear it from extra spaces and other elements.

```python
with open('C:\\Books\\result\\data\\basic_data_dict.txt', 'r') as basic_dict:
    for line in basic_dict.readlines():
        clear_line = line.strip()
        if len(clear_line) > 0:
            clear_data_word = clear_line.split(', ')[0].split(' ')[0].strip()
            clear_basic_data.append(clear_data_word)
            
```

The following check showed, that some of the words still have numbers, extra punctuaion and also not all of the item were correct in the dictionary in the first place – they have broken symbols, extra latinic symbols etc.

```python
print([word for word in clear_basic_data[0:150]], sep='\n')
```

And we decided to exclude elements mentioned above and also some extra affixation, which is not important for the main task. All the items also were formatted appropriately.

```python
clear_basic_data = list(map(lambda x: x.lower(), filter(lambda x: x.isupper() and x.isalpha(), clear_basic_data)))
```

```python
print(*[word for word in clear_basic_data[0:150]], sep='\n')
```

Extra latin entries were filtered with regular expressions. As we can see, there were nearly 700 broken elemnts (they can be restored from the other sources).

```python
r1 = re.compile('[а-яА-Я]+')
r2 = re.compile('[a-zA-Z]+')
filtered_clear_basic_data = sorted(list(set([word for word in filter(r1.match, clear_basic_data)])))
not_clear_basic_data = [word for word in filter(r2.match, clear_basic_data)]
print(*[word for word in filtered_clear_basic_data[0:150]], sep='\n')
```

For the main task we used function ‘check_the_word’, which takes two arguments: basic word (in our case it is a ‘клммуникация’) and any other word. We used the function on the clean dictionary.

```python
def check_the_word(basic_word, dict_item):
    if len(dict_item) <= len(basic_word):
        sorted_item = sorted(list(dict_item))
        sorted_word = sorted(list(basic_word))
        i, j = 0, 0
        while i < len(sorted_word) and j < len(sorted_item):
            if sorted_item[j] == sorted_word[i]:
                i += 1
                j += 1
            elif sorted_item[j] != sorted_word[i]:
                i += 1
        if j == len(sorted_item):
            return True

word = 'коммуникация'
result = [item for item in filtered_clear_basic_data if check_the_word(word, item)]
        
```

Obtained results has some extra elements, such as title letters, conjunctions, prepositions and flexions, so we exclude them.

```python
print(result)
print(len(result))
```
