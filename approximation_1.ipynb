# Main task: validate several approximation functions according to the received data

## **Case 1.** Data: (2; 12.258), (4; 17.24), (8; 30.151). Function: $f(x) = 3x + 5$

**1.1.** Putting down the function

```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
f = 3*x + 5
f
```

**1.2.** Getting MSE and RMSE value

```python
mse = ((12.258 - f.subs(x, 2))**2 + (17.24 - f.subs(x, 4))**2 + (30.151 - f.subs(x, 8))**2) / 3
mse
```

```python
rmse = mse**0.5
rmse
```

```python
plot(f)
```

**1.3.** Comparing actual and received data

```python
y_1 = f.subs(x, 2)
y_2 = f.subs(x, 4)
y_3 = f.subs(x, 8)
```

```python
y_1, y_2, y_3
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.solvers.inequalities import solve_univariate_inequality

x_range = np.linspace(-3, 10, num=100)
f = lambda x: 3*x + 5
plt.scatter([2, 4, 8], [11, 17, 29], c='r')
plt.scatter([2, 4, 8], [12.258, 17.24, 30.151], c='g')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.plot(x_range, f(x_range))
```

```python
data = abs((12.258 - 11)/12.258) + abs((17.24 - 17)/17.24) + abs((30.151 - 29)/30.151)
avg_apr_er = (data / 3) * 100
avg_apr_er
```

**Conclusion:** as average approximation error < 15% and root square middle error < 180, the proposed model $f(x) = 3x + 5$ can be used for this case.
